## Setup Repositories
Assuming the repositories are either cloned here or mounted via Google Drive.
For Colab, you might want to zip and upload your local `cosmic_engine`, `weorold`, and `worldmaker` folders, or mount your drive.

```python
from google.colab import drive
drive.mount('/content/drive')

# NOTE: Adjust these paths to where your repos are stored in Drive!
import sys
import os

import subprocess
repos = {
    'worldmaker': 'https://github.com/pmcray/worldmaker.git',
    'weorold': 'https://github.com/pmcray/weorold.git',
    'cosmic_engine': 'https://github.com/pmcray/cosmic_engine.git'
}
for repo, url in repos.items():
    path = f'/content/drive/MyDrive/{repo}'
    if not os.path.exists(path):
        print(f"Cloning {url} into {path}...")
        subprocess.run(['git', 'clone', url, path])
    if path not in sys.path:
        sys.path.append(path)
```
