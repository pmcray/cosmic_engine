```
import taichi as ti
import numpy as np
import matplotlib.pyplot as plt

# You may need to change arch=ti.vulkan or ti.cuda depending on Colab runtime
ti.init(arch=ti.cpu)

from render.camera import SphereCamera
from physics.fluid_solver import FluidEngine
from physics.nexus_core import NexusCore
from agi_operator import AGIOperator

res = 512
fluid = FluidEngine(res=res)
camera = SphereCamera(fluid_res=res, render_res=res, has_rings=0, samples=4)
operator = AGIOperator()

print("Pre-warming Jovian Fluid Dynamics...")
for i in range(100):
    fluid.step(i)
print("Warming Complete.")
```

```
def plot_frame(frame, attention, tilt, pan):
    # Set the Agent/Observer attention level
    operator.target_attention = attention
    operator.attention_level = attention
    camera.geo_engine.update_observer_attention(attention)
    
    camera.cam_tilt = tilt
    camera.cam_pan = pan
    
    fluid.step(100 + frame)
    
    # Render (light dir: lx=1.0, ly=0.5, lz=0.0)
    camera.render_gas_giant(fluid.dye, 1.0, 0.5, 0.0, frame * 0.016)
    img = camera.get_image_data()
    
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Frame: {frame} | Agent Attention: {attention:.2f}")
    plt.show()
```

## Phase 1: Unobserved State
The Operator is passive (attention=0.0). The Geodesic Engine saves compute by rendering the probabilistic Pi-Lattice wave skeleton instead of full raymarching.

```
plot_frame(0, attention=0.0, tilt=-0.45, pan=0.0)
```

## Phase 2: Interrogation
The Operator notices an anomaly and begins focusing (attention=0.5). ZPHC collapse blends the wave-skeleton with geometric reality.

```
plot_frame(20, attention=0.5, tilt=-0.35, pan=0.2)
```

## Phase 3: Total Collapse
The Operator is fully locked on (attention=1.0). The underlying lattice fully collapses into photorealistic planetary physics.

```
plot_frame(40, attention=1.0, tilt=-0.25, pan=0.4)
```
